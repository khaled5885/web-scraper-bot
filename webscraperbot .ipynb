```python
import selenium
from selenium import webdriver as wb
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
s=Service('D:\port\.venv\Include\chromedriver.exe')
driver=wb.Chrome(service=s)
driver.get('https://webscraper.io/test-sites/e-commerce/static')
time.sleep(2)
h=driver.find_element(By.XPATH,"""//*[@id="side-menu"]/li[2]""")
h.click()
time.sleep(1)
n=driver.find_element(By.XPATH,"""//*[@id="side-menu"]/li[2]/ul/li[1]""").click()
time.sleep(1)
v=driver.find_element(By.XPATH,"""/html/body/div[3]/div[2]""").click()
x=driver.find_elements(By.CLASS_NAME,'row')

```

```python
listoflinks = []
num_pages = 0
max_pages = 20
while num_pages < max_pages:
    st = driver.find_elements(By.CLASS_NAME, 'thumbnail')
    for element in st:
        s = element.find_elements(By.TAG_NAME, 'h4')[-1]
        s1 = s.find_element(By.TAG_NAME, 'a')
        listoflinks.append(s1.get_property('href'))

    try:
        next_button = driver.find_elements(By.CLASS_NAME, 'page-item')[-1]
        if next_button.is_enabled() and next_button.is_displayed():
            next_button.click()
            time.sleep(1)  # add a 1-second delay
            num_pages += 1
        else:
            break
    except NoSuchElementException:
        break
```

```python
from tqdm import tqdm
alldetails=[]
for i in tqdm(listoflinks):
    s=Service('D:\port\.venv\Include\chromedriver.exe')
    driver=wb.Chrome(service=s)
    driver.get(i)
    producte=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div/div[2]/div/div/div/div[2]/div[1]/h4[2]').text
    price=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div/div[2]/div/div/div/div[2]/div[1]/h4[1]').text
    review=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div/div[2]/div/div/div/div[2]/div[3]/p').text
    description=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div/div[2]/div/div/div/div[2]/div[1]/p').text
    items={'producte':producte,
       'price':price,
       'review':review,
       'description':description,
       'link':'i'}
    alldetails.append(items)
```

```python
import pandas as pd
pd.DataFrame(alldetails)
```
